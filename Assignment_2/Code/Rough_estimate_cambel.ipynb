```python
import numpy as np
from matplotlib import pyplot as plt

```

```python
Blade_1_st_flap = 0.63 * 60  #RPM
Blade_1_st_edge = 0.96 * 60#RPM
Tower = 0.25 * 60 #RPM

rpm_rotor = np.arange(0,9.6,0.1)
```

```python
"""BW modes decrease with ùú¥
FW modes increase with ùú¥"""

BW_1_st_flap = Blade_1_st_flap - rpm_rotor
FW_1_st_flap = Blade_1_st_flap + rpm_rotor
SYM_1st_flap = np.ones(len(rpm_rotor))*Blade_1_st_flap

BW_1_st_edge = Blade_1_st_edge - rpm_rotor
FW_1_st_edge = Blade_1_st_edge + rpm_rotor
SYM_1st_edge = np.ones(len(rpm_rotor))*Blade_1_st_edge
```

```python
plt.plot(rpm_rotor,BW_1_st_flap/60,label='BW Blade 1st flap')
plt.plot(rpm_rotor,FW_1_st_flap/60,label='FW Blade 1st flap')
plt.plot(rpm_rotor,SYM_1st_flap/60,label='SYM 1st flap')
plt.plot(rpm_rotor,SYM_1st_edge/60,label='SYM 1st edge')
plt.plot(rpm_rotor,BW_1_st_edge/60,label='BW Blade 1st edge')
plt.plot(rpm_rotor,FW_1_st_edge/60,label='FW Blade 1st edge')
plt.plot(rpm_rotor,Tower*np.ones(len(rpm_rotor))/60,label='Tower')

plt.legend(loc='upper right')
plt.xlabel('Rotor rotational speed $\Omega$ (RPM)')
plt.ylabel('Natural Frequency (Hz)')
plt.show()
```
